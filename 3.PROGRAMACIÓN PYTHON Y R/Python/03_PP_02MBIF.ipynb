# Ficheros

La mayoría de las veces los datos se obtienen de un fichero. Para leer un fichero hay que indicarle a Python que abra ese fichero. La operación se realizar con la instrucción `open()`

```python
# Abriendo un fichero
f = open('10_sequences.seq', 'r')
type(f)
```

```python
# Leyendo líneas del fichero
f.readline()
```

```python
#  Leyendo todas las líneas del fichero.readline()
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
print(f.readline())
```

```python
#  Mejor si añadimos un bucle
f = open('10_sequences.seq', 'r')
for linea in f:
    print(linea)
```

```python
# Una tarea algo más compleja.
# Listar todas las secuencias que empiecen por citosina
# Método nuevo `strip()`
f = open('10_sequences.seq', 'r')
for linea in f:
    linea = linea.strip()
    if linea.startswith('C'):
        print(linea)
```

## Funciones
Existen funciones integradas en Python y funciones definidas por el usuario. Una función es un bloque de código con un nombre.

```python
# Función integrada en Python
len('AGGTTCATGAAACCCC')
```

Una función siempre empieza con la palabra reservada del lenguaje `def`, luego el nombre de la función y a continuación la lista de parámetros. La función puede no tener parámetros. Luego viene un bloque de código.

```python
# Función definida por el usuario
def bienvenida_curso():
    print('Bienvenidos al curso de programación con Python y R de la VIU!!')
```

```python
# Ejecutando la función 
bienvenida_curso()
```

```python
bienvenida_curso()
```

Los argumentos y parámetros permiten usar el algoritmo implementado en la función con distintos datos.

```python
# Función definida por el usuario
def bienvenida_curso(nombre):
    print('Hola {}. Genial de contar contigo en el curso de programación con Python y R de la VIU!!'.format(nombre))
```

```python
bienvenida_curso()
```

```python
# Ejecución de la función
bienvenida_curso('Juan')
```

La sentencia `return` le indica a Python que debe salir de la función y devolver un resultado. El resultado puede ser cualquier tipo de objeto.

```python
# Las funciones pueden tener múltiples parámetros y, en general, devuelven un resultado
def resta(x,y):
    resta = x-y
    return resta
```

```python
# Ejecutando la función resta()
resta(19, 73)
```

```python
valor_resta = resta(19,73)
```

```python
valor_resta
```
