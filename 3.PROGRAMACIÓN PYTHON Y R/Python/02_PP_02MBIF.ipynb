# Listas y el bucle for

Las listas son secuencias ordenadas de objetos. Por ejemplo, una lista con sitios de restricción* sería ['GAATTC', 'GGATCC', 'AAGCTT'].

*Los sitios de restricción son secuencias de ADN que son reconocida por una endonucleasa (una proteína que corta el ADN) como el lugar en el que se cortará el ADN, o también llamado "diana de corte".

```python
# Crear una lista vacia
sitios_restriccion = []
print(sitios_restriccion)
```

```python
#  Añadir un elemento a la lista
sitios_restriccion.append('GAATTC')
print(sitios_restriccion)
```

```python
#  Añadir varios elementos a la lista
sitios_restriccion.extend(['GGATCC', 'AAGCTT'])
print(sitios_restriccion)
```

## El bucle for
Se utiliza cuando se quiere hacer algo a cada elemento de una lista. Los elementos que los componen:
- for *variable* in [lista]: 
    - Nuevo bloque de código (el que se va a repetir. Tiene que estar identado)

```python
# Ejemplo de bucle for
proteina = ['ALA', 'PRO', 'ILU', 'CYS']
for aminoacido in proteina:
      print('Aminoácido: ', aminoacido)
```

```python
# Bucle for con varias instrucciones
secuencias = ['GAATTCAC', 'GGCATCACC', 'AAGCTTACTGC']
for secuencia in secuencias:
    print('Secuencias: ', secuencia)
    print('Número de bases: ', len(secuencia))
```

```python
# Bucle for con varias instrucciones e identación incorrecta
secuencias = ['GAATTCAC', 'GGCATCACC', 'AAGCTTACTGC']
for secuencia in secuencias:
    print('Secuencias: ', secuencia)
        print('Número de bases: ', len(secuencia))
```

```python
# El bucle for también funciona para strings
secuencia =  'AAGCTTACTGC'
for base in secuencia:
    print('Base: ', base)
```

```python
# Numerando las bases
seq = "AGTCATGTCGC"
n = 1
for base in seq:
    print('La base {} es {}'.format(n, base))
    n = n+1
```

## La sentencia if
Ejecuta un bloque de código sólo cuando algo es verdad.

```python
"GAATTC" in "ATCTGGAATCATCG"
```

```python
# Ejemplo if (si se cumple condición ejecuto bloque de código)
if "GAATTC" in "ATCTGGAATCATCG":
    print("Sitio EcoRI está presente")
```

```python
# Si queremos que el bloque de código se ejecute cuando la condición no se cumpla
# Opción 1: operador not in
if "GAATTC" not in "AAAAAAAAA":
    print("EcoRI no cortará la secuencia")
```

```python
# Si queremos que el bloque de código se ejecute cuando la condición no se cumpla
# Opción 2: operador not cambia True por False
if not "GAATTC" in "AAAAAAAAA":
    print("EcoRI no cortará la secuencia")
```

### ¿Dónde está el sitio?

```python
# Recordemos el método find()
seq = "ATCTGGAATTCATCG"
seq.find("GAATTC")
```

```python
# Recordemos el método find() cuando la subsecuencia no existe
seq = "ATCTGGAATTCATCG"
seq.find("GGCGC")
```

```python
# Localizando el sitio de restricción
seq = "ATCTGGAATTCATCG"
pos = seq.find("GAATTC")
if pos == -1:
    print("EcoRI no cortará la secuencia")
else:
    print("El sitio EcoRI empieza en el índice: ", pos)
```

### Un ejemplo más complejo

```python
# Utilizando if dentro de un bucle for
sitios_restriccion = [ "GAATTC", "GGATCC", "AAGCTT"]
seq = input("Introduce una secuencia de ADN: ")
for sitio in sitios_restriccion:
    if sitio in seq:
        print("{} es un sitio de corte".format(sitio))
    else:
        print("No esá presente el sitio {}.".format(sitio))
```
