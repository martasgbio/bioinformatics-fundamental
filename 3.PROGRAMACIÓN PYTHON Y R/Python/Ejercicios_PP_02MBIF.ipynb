```python
# Ejercicio 2
# Implementar una función que recibe una secuencia de ADN de entrada y cuenta cúantas bases de cada tipo hay en la secuencia. 
# La función debe devolver una lista de tuplas (x,y). El primer elemento de la tupla será la letra de la base 
# y el segundo elemento de la tupla el número de bases que hay de esa letra.

# Definimos la función
def contador_bases(secuencia):

    # Definimos una lista vacía para almacenar las tuplas
    lista = []

    # Recorremos la secuencia base a base
    for base in secuencia:

        # Genero una tupla con primer elemento la base y segundo el número de apariciones de la base
        tupla = (base, secuencia.count(base))
        
        # Verifico si la tupla está en mi lista. Si no está la añado
        if tupla not in lista:
            lista.append(tupla)

    # Devuelvo la lista como salida de la función
    return lista
```

```python
# Pruebo la función con dos casos de prueba
base1 = 'TCTGAAAAGTGCAAAGAACGATGATGATGATGATAGAGGAACCTGAGCAGCCATGTCTGAACCTATAGC'
base2 = 'CCACACCAAAAAAACTTTCCACGTGTGAACTATACTCCAAAAACGAAGTATTGGTTTATCATAA'

# Ejecución de la función con la base 1
print('Ejercición de la función con el caso de prueba 1: ',contador_bases(base1))

# Ejecución de la función con la base 2
print('Ejercición de la función con el caso de prueba 2: ',contador_bases(base2))
```

```python
# Ejercicio 3
# Implementar una función que recibe una secuencia de ADN de entrada y la divide en codones.
# La función devuelve una lista con los codones.

# Definición de la función
def division_codones(DNA_seq):

    # Definimos lista vacía para almacenar los codones
    lista_codones = []

    # Bucle para ir generando los codones. El contador sala de 3 en 3 elementos. El cálculo del resto mediante
    # el operdor módulo (%) nos permite considerr los casos que no son múltiplo de 3. 
    for i in range(0, len(DNA_seq)-len(DNA_seq)%3, 3):

        # Selección con slicing los codones
        codones = DNA_seq[i:i+3] 
        
        # Añado los codones a la lista a generar
        lista_codones.append(codones)

    # La función genera como salida la lista de codones
    return lista_codones

```

```python
# Pruebo la función con tres casos de prueba: secuencia multiplo de tres, resto 2 y resto 1
base1 = 'CCACACCAAAAAAACTTTCCA'
base2 = 'CCACACCAAAAAAACTTTCA'
base3 = 'CCACACCAAAAAAACTTTC'

# Ejecución de la función con la base 1
print('Caso de prueba:  secuencia múltiplo de 3: ',division_codones(base1))

# Ejecución de la función con la base 2
print('Caso de prueba:  secuencia no multiplo de 3 (resto 1): ',division_codones(base2))

# Ejecución de la función con la base 2
print('Caso de prueba:  secuencia no multiplo de 3 (resto 2): ',division_codones(base3))
```

### Ejercicio 3
Lee el fichero *sequences.seq*.
- A. ¿Cuántas secuencias hay en ese archivo?
- B. ¿Cuántos tienen el patrón CTATA?

- D. ¿Cuántos tienen más del 50 % de composición de GC?
- E. ¿Cuántos tienen más de 2000 bases y más del 50% de composición de GC?
- F. ¿Sólo hay A, C, G, T? Muestra qué otras letras hay en las secuencias. Sólo mostrar cada letra una vez.

```python
# lectura de fichero sequences.seq

# Contadores para los distintos cálculos que solicita el ejercicio
# ################################################################

# Total de secuencias
tot_lineas = 0

# Secuencias con patrón CTATA
tot_lineas_con_patron = 0

# Secuencias con más de 1000 bases
tot_lineas1000 = 0

# Secuencias con más del 50% commposicion GC
tot_lineas50 = 0

# Secuencias que cumplen la doble condición: más de 2000 bases y más del 50% composicion GC
tot_lineas_doble_condi = 0

# Definimos una lista con las bases a considerar y otra para almacenar otras letras que pueda haber en la secuencia
bases = ["A", "C", "G", "T"]
otras_bases = []

# Proceso a realizar

# Abrimos el fichero en modo lectura
f = open('sequences.seq', 'r')

# Bucle for que lee las secuencias del fichero
for linea in f:

    # Contador de líneas
    tot_lineas +=1         
    
    # Eliminamos espacios en blanco
    linea = linea.strip()               

    #Verificamos si está patrón y en caso afirmativo incrementamos contador
    if "CTATA" in linea:           
        tot_lineas_con_patron += 1               

         
    # Verificamos si la secuencia tiene más de 1000 bases y en caso afirmativo incrementamos el contador
    if len(linea) > 1000:
        tot_lineas1000 += 1

    
    # Calculamos la composicion GC
    if "A" in linea:

                conteoA = linea.count("A") 

    if "C" in linea:

                conteoC = linea.count("C") 

    if "G" in linea:

                conteoG = linea.count("G") 

    if "T" in linea:

                conteoT = linea.count("T")

    porcentaje_GC = (((conteoC + conteoG)/(conteoA + conteoG + conteoC + conteoT))*100)         #calcula el porcentaje

    # Verificamos si la composición GC es superior al 50% y en caso afirmativo incrementamos el contador
    if porcentaje_GC > 50:
        tot_lineas50 += 1

        
    # Verificamos la doble condisión: más de 2000 bases y proporción GC mayor al 50%
    if len(linea) > 2000 and porcentaje_GC > 50:
        tot_lineas_doble_condi += 1


    # Verificamos si existen otras letras distintas a los cuatro nucleótidos
    for otra_letra in linea:
        if otra_letra not in bases and otra_letra not in otras_bases:
              otras_bases.append(otra_letra)

# Mostramos por pantalla los resultados
print("El total de secuencias es: ", tot_lineas)

print("El número de secuencias que contienen el patrón es: ", tot_lineas_con_patron)

print("El número de secuencias con más de mil bases es:", tot_lineas1000)

print("El número de secuencias con porcentaje de GC mayor que 50 es:", tot_lineas50)

print("El total de secuencias que cumple las dos condiciones son:", tot_lineas_doble_condi)   

print("Las otras letras en la secuencia son: " + ", ".join(otras_bases) +".")
```

### Ejercicio 6
Escriba un programa que imprima el complemento inverso de cada secuencia del archivo *10_sequences.seq*.
Este archivo contiene solo letras A, T, C y G.

```python
# Código Python
f = open('10_sequences.seq', 'r')

for sec in f:
    # Calculamos el complementos con los métodos maketrans() y translate()
    complementos = sec.maketrans('acgtACGT', 'tgcaTGCA')
    complement_sec = sec.translate(complementos)
    
    # Calculamos el string invertido
    complement_sec_rev = complement_sec[::-1]
    
    # Visualizamos la salida
    print('Complemento reverso: ', complement_sec_rev)
```

Escriba un programa para pedir una secuencia de ADN. Traducir el ADN en proteína. (Vea la celda siguiente para la tabla de codones a usar). Cuando el codón no está codificado en la tabla (p. ej., codón de terminación), use \*. Ignorar las bases extra si la longitud de la secuencia no es múltiplo de 3. Decide cómo quieres manejar los códigos ambiguos.

```python
# Convesión codón a proteina
table = {
 'TTT': 'F', 'TTC': 'F', 'TTA': 'L', 'TTG': 'L', 'TCT': 'S',
 'TCC': 'S', 'TCA': 'S', 'TCG': 'S', 'TAT': 'Y', 'TAC': 'Y',
 'TGT': 'C', 'TGC': 'C', 'TGG': 'W', 'CTT': 'L', 'CTC': 'L',
 'CTA': 'L', 'CTG': 'L', 'CCT': 'P', 'CCC': 'P', 'CCA': 'P',
 'CCG': 'P', 'CAT': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
 'CGT': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R', 'ATT': 'I',
 'ATC': 'I', 'ATA': 'I', 'ATG': 'M', 'ACT': 'T', 'ACC': 'T',
 'ACA': 'T', 'ACG': 'T', 'AAT': 'N', 'AAC': 'N', 'AAA': 'K',
 'AAG': 'K', 'AGT': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
 'GTT': 'V', 'GTC': 'V', 'GTA': 'V', 'GTG': 'V', 'GCT': 'A',
 'GCC': 'A', 'GCA': 'A', 'GCG': 'A', 'GAT': 'D', 'GAC': 'D',
 'GAA': 'E', 'GAG': 'E', 'GGT': 'G', 'GGC': 'G', 'GGA': 'G',
 'GGG': 'G', }
# Datos extra
stop_codons = [ 'TAA', 'TAG', 'TGA']
start_codons = [ 'TTG', 'CTG', 'ATG']
```

```python
# Pedir una secuencia de ADN y agruparla en tripletes:
secuencia = input('Introduzca una secuencia de ADN: ').upper()
tripletes = []
aminoacidos = []

# Generamos tripletes
for i in range(0,len(secuencia) - len(secuencia)%3, 3):
    tripletes.append(secuencia[i:i+3])
    
#Traducir los tripletes a aminoácidos:
for base in tripletes:
    table[base]=table.get(base,'*')
    aminoacidos_traducidos = table[base]
    aminoacidos.append(aminoacidos_traducidos)

proteina = ''.join(aminoacidos)
        
print('Los aminoácidos codificados son: ',proteina)
```

###  Ejercicio 1 Buscando patrones

Verifica si el patrón motif = r'ATG.*?TAA' esta presente en la secuencia 'AATGAAGGGCCGCTACGATAAGGAACTTCGTAATTTCAG'. Si está imprime la posición inicial y final en la que aparece el patrón en la secuencia.

```python
# Codigo Python

# importamos módulo re
import re

# Definimos patrón
pattern = r"ATG.*?TAA"

# Secuencia
seq = "AATGAAGGGCCGCTACGATAAGGAACTTCGTAATTTCAG"

# Buscamos las ocurrencias del patrón
match = re.search(pattern, seq)

if match:
    print('Encontrado el patrón en la secuencia: ', seq)
    print('Inicio patrón: ', match.start())
    print('Fin patrón: ', match.end())    
else:
    print(seq, "No encontrado")
```
