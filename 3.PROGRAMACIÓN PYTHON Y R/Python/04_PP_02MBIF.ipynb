### Códigos de ambigüedad
A veces, las bases del ADN son ambiguas. Por ejemplo, el secuenciador podría decir que una base no es una G o una T, pero podría ser una A o una C. El código estándar de una letra (IUPAC) para el ADN incluye letras para la ambigüedad:

- M es A o C
- R es A o G
- W es A o T
- S es C o G
- Y es C o T
- K es G o T
- V es A, C o G
- H es A, C o T
- D es A, G o T
- B es C, G o T
- N es G,A,T o C

```python
# Contando bases
seq = "TKKAMRCRAATARKWC"
cuenta = {}
for letra in "ABCDGHKMNRSTVWY":
    cuenta[letra] = seq.count(letra)

# Visualizamos cuenta
print(cuenta)
```

```python
len(cuenta)
```

```python
# Visualizamos cuenta
for base in cuenta:
    print(base, "=" , cuenta[base])
```

```python
# Contando bases pero no sabemos las bases a priori
seq = "KTKAMRCRAATARKWC"
cuenta = {}
for base in seq:
    if base not in cuenta:
        cuenta[base] = 1
    else:
        cuenta[base] = cuenta[base] + 1
        #cuenta[base]+=1
```

```python
# Visualizamos cuenta
print(cuenta)
```

```python
# Contando bases pero no sabemos las bases a priori. Usando get
seq = "TKKAMRCRAATARKWC"
cuenta = {}
for base in seq:
    cuenta[base] = cuenta.get(base, 0) + 1
```

```python
# Visualizamos cuenta
print(cuenta)
```

```python
### Cadena complementaria reversa de una secuencia I
tabla_complementos = {"A": "T", "T": "A", "C": "G", "G": "C"}

seq = "CTCGTATT"
new_seq = []
for letra in seq:
    letra_complemento = tabla_complementos[letra]
    new_seq.append(letra_complemento)
    
print(new_seq)
```

```python
new_seq
```

```python
### Complemento inverso de una secuencia II
new_seq.reverse()
print(new_seq)
```

```python
### Complemento inverso de una secuencia III
new_seq = ''.join(new_seq)
print(new_seq)
```

```python
secuencia_revertida = new_seq[::-1]
```

```python
secuencia_revertida
```

```python
# Contando codones
seq = "TCTCCAAGACGCATCCCAGTGA"
cuenta_codon = {}
for i in range(0, len(seq) - len(seq)%3, 3):
    codon = seq[i:i+3]
    cuenta_codon[codon] = cuenta_codon.get(codon, 0) + 1
print(cuenta_codon)
```

```python
# Ordenando los codones
codones = list(cuenta_codon.keys())
print(codones)
codones.sort()
print(codones)
```

```python
# Métdoo maketrans() y translate() de un string. Permite construir una tabla de traducción y hacer la traducción.
zika_DNA = 'AGTTGTTGATCTGTGTGAGTCAG'
 
complements = zika_DNA.maketrans('acgtACGT', 'tgcaTGCA')
complement_seq = zika_DNA.translate(complements)
print(complement_seq)
```

```python
# If-elif
DNA_segment = 'ATGACATGACCAATC'
codon1 = DNA_segment[-3:]
 
if (codon1 == 'ATG'):
    print('Codon', codon1, 'Es un codón de inicio.')
elif ((codon1 == 'TAA') or 
     (codon1 == 'TAG') or
     (codon1 == 'TGA')):
    print('Codon', codon1, 'no es ni un codón de inicio ni de fin.')
elif (codon1 == 'ATA'):
    print('Codon', codon1, 'no es un codón de inicio ni de fin.')
else:
    print('El codón', codon1, 'es un codón de fin.')
```

```python
# Bucle while itera el bloque de instrucciones siempre que EXPR siga siendo True.
DNA_seq = 'CGGCAACACAAAAAGAATGAAGGATTTTGAATCTTTATTGTGTGCGAGTAACTACGAGGAAGATTAAAGA'
print('Secuencia DNA:', DNA_seq)
 
bp = 'T'
print('Base:', bp)
 
print('str.count():', DNA_seq.count(bp))
 
count = 0
index = 0
 
while index < len(DNA_seq):
    if bp == DNA_seq[index]:
        count = count + 1
    index += 1
 
print('Cuenta con while:', count)
```
