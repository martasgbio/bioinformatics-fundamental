## Creando strings
Los strings comienzan o acaban con una comilla doble o simple (debe ser la misma):
- "Esto es un string"
- 'Esto es otro string'
- "Da igual si usas comilla simple \' o comilla doble \""

La barra invertida (backslash) \ se utiliza para representar caracteres especiales en un string.

```python
# Ejemplos de carácter especial
s = "Da igual si usas comilla simple \' o comilla doble \""
print(s)
```

```python
# Mas caracteres especiales
s = 'Los caracteteres especiales se introducen en un string con la barra invertida \\'
print(s)
```

```python
# Mas caracteres especiales
s = 'La nueva línea se introduce con la secuencia de escape \\n '
print(s)
```

```python
# Ejemplo de nueva línea
s = 'Aquí vemos \\n un salto de linea '
print(s)
```

```python
# Ejemplo de tabulación
s = 'Aquí vemos \t un tabulado '
print(s)
```

## ¿Cambiando un string?
Los strings no se pueden modificar. Son inmutables. Si queremos cambiarlo hay que crear uno nuevo.

```python
# Intentando modificar un string
s = 'GATTACA'
s[3]='C'
```

```python
# Creando uno nuevo
s = s[:3] + 'C' + s[4:]
s
```
